## Problem 1

College juniors (undergraduates) at US universities were asked if they were “unlikely”, “somewhat likely”, or “very likely” to apply to graduate school. Data on whether the parents have graduate education (pared), whether the undergraduate institution is public (public), and current GPA (gpa), were also collected.

a) Fit a multinomial model to predict the probabilities for graduate application by category, “unlikely”, “somewhat likely”, or “very likely”.

```r
dat=read.csv('res/dat.csv')
dat$X=NULL
```

```r
library(nnet)

#transform the labels
levels(dat$apply)
#table(dat$pared, dat$apply)
#table(dat$public,dat$apply)

# fit the model
mmod=multinom(apply ~ as.numeric(pared) + as.numeric(public) + as.numeric(gpa),data = dat)
summary(mmod)
```

b). Repeat the analysis with an ordinal model. Comment on any differences.

```r
library(MASS)
omod <- polr(apply ~ pared+public+gpa, data=dat) 
summary(omod)
```

```r
print(mmod$edf)
print(omod$edf)
```

### Comments

In addition, there multinomial has 8 df and 5 d.f in the ordinal model since we estimated less parameters.

c) Write an R function to implement (b). Then use it to generate a sample of 1000 numbers from t(3), and compare the sample pdf curve with the actual t(3) curve.

```r
cauchy_sim <- function(n) {
  sqrt(3)*tan(pi*runif(n, -0.5, 0.5))
}

t_sim <- function() {
  # simulate a standard normal
  c <- 2
  while (TRUE) {
    x <- cauchy_sim(1)
    y <- runif(1, 0, c*sqrt(3)/(3 + x^2)/pi) 
    if (y < dt(x,3)){
      return(x)  
    } 
  }
}

set.seed(20)
n <- 1000
x <- rep(0, n)
for (i in 1:n) x[i] <- t_sim()

hist(x, freq=FALSE,ylim=c(0,0.5),breaks = 30)
xt <- seq(-15,15,0.01)
lines(xt,dt(xt,3),col="blue")
curve(2/(sqrt(3)*pi)*(1+((x^2)/3))^(-2), add=TRUE, col="red")
```

## Problem 4

There is a famous data set from RA Fisher’s 1936 paper on The use of multiple measurements in taxonomic problems which consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimetres.

The data is available in R in the data set iris.

Perform a principal components and cluster analysis on the four variables using the R commands
prcomp and kmeans. Use two separate methods to determine the number of clusters.
Use the ggplot2 and ggfortify libraries to do plots of the first two principal components that
indicate where the directions of the original variables and the positions of the cluster means. Comment on your results.

```r
# import libraries
library(ggplot2)
library(ggfortify)
library(factoextra)
library(NbClust)
library(cluster)
```

First, a little bit of exploration:

```r
# load and view the dataset
data(iris)
str(iris) 
summary(iris[1:4])

# plot the dataset given the species
pairs(iris[1:4],col=as.numeric(iris$Species)+1)
```

```r
# standarize the values
iris.s = data.frame(scale(iris[,1:4]))

# calculate pca
pca =prcomp(iris.s,scale=T)
summary(pca)
```

This gives us the standard deviation of each component, and the proportion of variance 
explained by each component.

It is shown that the first two principal components account for most of the variation 
in all four components. Specifically, the first two PC account for 95.81% of the variation. The first component is explaining most of the variance (72.96%).

```r
# the matrix of variable loadings:
pca$rotation
```

The component weightings show that PC1 put similar weights the length (for both, sepal and petal) and also the petal width. The second component, PC2, is setting bigger weights on sepal (both length and width) and a significant small weight for petal.

```r
# colour by species
autoplot(pca,data= iris,colour="Species",loadings = TRUE, loadings.label = TRUE) 
```

We can plot the variance to see how many cluster we can define for kmeans

```r
plot(pca, type="lines")
```

Regarding to clustering, the method suggests to use k=3 clusters for our model. The next figure shows kmeans model with k=3

```r
k.max <- 15 # Maximal number of clusters

wss <- sapply(1:k.max,function(k){
  kmeans(iris.s, k, nstart=10 )$tot.withinss
  }) 
plot(1:k.max, wss,type="b", pch = 19, frame = FALSE, xlab="Number of clusters K", ylab="Total within-clusters sum of squares")
abline(v = 3, lty =2)
```

```r
k.iris = kmeans(iris.s,3) 
#autoplot(k.iris,data=iris.s)
fviz_cluster(k.iris, geom = "point", data = iris.s)
```

The next figure shows the silhouette method to analyse and verify the number of clusters for kmeans.

```r
# try to draw centers
#points(k.iris$centers, pch = 8, cex = 2)

dis = dist(iris.s)^2
sil = silhouette (k.iris$cluster, dis) 

fviz_nbclust(iris.s, kmeans, method = c("silhouette"))
```

Silhouette method suggests k=2 as number of clusters.

```r
kmeans_2 <- kmeans(iris.s, 2)
fviz_cluster(kmeans_2, geom = "point", data = iris.s)
```

It's known the number of cluster should be 3 to match the number of classes. However, is interesting how this two methods present different results which can benefit further analysis where we actually don't know the answer and we are applying unsupervised learning.
