## Problem 1

Suppose that $A$ is a symmetric matrix with $A^k = A^{k+1}$ for some integer $k ≥ 1$. Show that $A$ is idempotent.


If $A$ is symmetric, then an orthogonal matrix $P$ exists that $\Lambda=P^TAP$ and $A=P\Lambda P^T$.<br>
Given that $A^k=A^{k+1}$, then<br>
$(P\Lambda P^T)^k=(P\Lambda^{k+1} P^T)$<br>
$(P\Lambda P^T)(P\Lambda P^T)...(P\Lambda P^T)=(P\Lambda P^T)(P\Lambda P^T)(P\Lambda P^T)...(P\Lambda P^T)$<br>
using asociative property and given that P is orthogonal then, P^TP=I, <br>
$P\Lambda (P^TP)\Lambda (P^TP)... \Lambda P^T=P\Lambda (P^TP)\Lambda (P^TP) \Lambda (P^TP)...\Lambda P^T$<br>
Multiplying the diagonal matrix $\Lambda $<br>
$\Lambda^k=\Lambda^{k+1}$<br>
This is only possible if every term in the diagonal of matrix $\Lambda$,  $\lambda_i^k=\lambda_i^{k+1}$, $i=1...dim(A)$<br>
Then,<br>
$\Lambda^k-\Lambda^{k+1}=0$<br>
$\Lambda^k(1-\Lambda^k)=0$<br>
The equation is satisfied by $\Lambda^k=0$ or $\Lambda^k=1$, we can conclude for $\Lambda$, the eigen values are $0$ or $1$ as $\Lambda$ can take only $0$ or $1$, then it is symmetric and idempotent.<br>
Since $\Lambda=P^TAP$, multiplying for $P$ and $P^T$, and given that $P^TP=PP^T=I$, then if $A=P\Lambda P^T$<br>
$A^k=A^{k+1}$<br>
$P\Lambda^k P^T=P\Lambda^{k+1} P^T$<br>
Replacing $\Lambda$ in both sencarios, $i)\Lambda = 0$ and $ii)\Lambda=1$, we will have the same result:
$P\Lambda^k P^T=P\Lambda^{k+1} P^T$, and as $\Lambda=P^TAP$<br>
we can conclude for any $k\ge 0$, and in particular, for $k=2$<br>
$A^2=A$ <b>, A is idempotent as required</b>


## Problem 2

Let $A_1, A_2, . . . , A_m$ be a set of symmetric $k × k$ matrices. Suppose that there exists an orthogonal matrix $P$ such that $P^TA_iP$ is diagonal for all $i$. Show that $A_iA_j = A_jA_i$ for every pair $i,j = 1,2,...,m.$

Let <br>
$\Lambda_i=P^TA_iP$ be the diagonal matrix for $A_i$ [1]<br>
$\Lambda_j=P^TA_jP$ be the diagonal matrix for $A_j$ [2] <br>

then, <br>
$A_i=P\Lambda_iP^T$<br>
$A_j=P\Lambda_jP^T$<br>

Now,
$A_iA_j=P\Lambda_iP^TP\Lambda_jP^T$, Given that $P$ is orthogonal, then $P^TP=I$<br>
$A_iA_j=P\Lambda_i\Lambda_jP^T$<br>
By definition, by asociative property and for symmetric diagonal matrix multiplication :
$diag(a_1, ..., a_n) · diag(b_1, ..., b_n) = diag(a_1b_1, ..., a_nb_n)=diag(b_1a_1, ..., b_na_n)$. Then,<br>
$A_iA_j=P(\Lambda_i\Lambda_j)P^T=P\Lambda_j\Lambda_iP^T$<br> 
<br>
Replacing [1] and [2]<br>
$A_iA_j=PP^TA_jPP^TA_iPP^T$<br>
using orthogonal $P$, then<br>
$A_iA_j=A_jA_i$ <b>as required.</b>


## Problem 3

Show that for any random vector $y$ and compatible matrix $A$, we have $var \ A \mathbf y = A(var \ \mathbf y)A^T$ 

From definition of variance of a random vector $var \ \mathbf y$,<br>
$Var \ \mathbf y=E[(\mathbf y- \boldsymbol \mu)(\mathbf y-\boldsymbol \mu)^T]$ [1], where $\boldsymbol \mu=E(\mathbf y)$<br>
$Var \ A\mathbf y=E[(A\mathbf y-A \boldsymbol \mu)(A\mathbf y-A \boldsymbol \mu)^T]$, assuming matrix $A$ is constant, we can get common factor<br>
$Var \ A\mathbf y=E[A(\mathbf y-\boldsymbol \mu)(A(\mathbf y-\boldsymbol \mu))^T]$, given that $(XY)^T=Y^TX^T$, then<br>
$Var \ A\mathbf y=E[A(\mathbf y-\boldsymbol \mu)(\mathbf y-\boldsymbol \mu)^TA^T]$, since $A$ and $A^T$ are constant, $E(A\mathbf x)=AE(\mathbf x)$ and $E(\mathbf x A^T)=E(\mathbf x)A^T$ for any vector $\mathbf x$, then: <br>
$Var \ A\mathbf y=AE[(\mathbf y-\boldsymbol \mu)(\mathbf y-\boldsymbol \mu)^T]A^T$, from [1],<br>
$Var \ A\mathbf y=AVar(\mathbf y)A^T$, <b>as required</b>

The table below shows prices in US cents per pound received by fishermen and vessel owners for various species of fish and shellfish in 1970 and 1980. (Taken from Moore & McCabe, Introduction to the Practice of Statistics, 1989.)

<b>a)</b>The linear model is of the form $y = X \boldsymbol \beta + \boldsymbol \epsilon$. Write down the matrices and vectors involved in this equation.

$y=\left[ \begin{array}{cccc}
27.3 \\ 42.4 \\ 38.7 \\ 4.5 \\ 23.0 \\ 166.3 \\ 109.7 \\ 80.1 \\ 150.7 \\ 20.3 \\ 189.7 \\ 131.3 \\ 404.2 \\ 149.0
\end{array} \right]$,
$X=\left[ \begin{array}{cccc}
1 & 13.1 \\ 1 &  15.3 \\ 1 &  25.8 \\ 1 &  1.8 \\ 1 &  4.9 \\ 1 &  55.4 \\ 1 &  39.3 \\ 1 &  26.7 \\ 1 &  47.5 \\ 1 &  6.6 \\ 1 &  94.7 \\ 1 &  61.1 \\ 1 &  135.6 \\ 1 &  47.6 \\
\end{array} \right]$,
$\beta=\left[ \begin{array}{cccc}
\beta_0  \\
\beta_1 \\
\end{array} \right]$,
$\epsilon=\left[ \begin{array}{cccc}
\epsilon_0 \\ \epsilon_1 \\ \epsilon_2\\ \epsilon_3 \\ \epsilon_4 \\ \epsilon_5 \\ \epsilon_6 \\ \epsilon_7 \\ \epsilon_8 \\ \epsilon_9 \\ \epsilon_{10} \\ \epsilon_{11} \\ \epsilon_{12} \\ \epsilon_{13}
\end{array} \right]$<br>

```r
price_1970=c(13.1, 15.3, 25.8, 1.8, 4.9, 55.4, 39.3, 26.7, 47.5, 6.6, 94.7, 61.1, 135.6, 47.6)
y=c(27.3, 42.4, 38.7, 4.5, 23.0, 166.3, 109.7, 80.1, 150.7, 20.3, 189.7, 131.3, 404.2, 149.0)
beta0=1
```

```r
X=cbind(beta0,price_1970)
#X=matrix(c(rep(1,14),price_1970),14,2)
X
```

Moreover, we can create the model in R:

```r
df_price=data.frame(y,price_1970)
model=lm(y ~ price_1970, data=df_price)
```

<b>c)</b> Calculate the sample variance $s^2$.

The sample variance,<br>
$s^2=\frac{(y-Xb)^T(y-Xb)}{n-p}$
is un unbiased estimator for $\sigma^2$<br>
where $p=2$ is the number of parameters in the model.

We can calculate the sum of squares of residuals $SS_{Res}=(\boldsymbol y-X \boldsymbol b)^T(\boldsymbol y-X \boldsymbol b)$<br>

```r
n=dim(X)[1]
e=y-X%*%b # alternative - model$residuals
SSres=sum(e^2) # alternative - (t(y-X%*%b)%*%(y-X%*%b))
s2=SSres/(dim(X)[1]-(length(b)))
s2
```

$s^2=777.1527$

As an alternative, we can use the R model to calculate the sample variance:

```r
deviance(model)/model$df.residual
```

```r
k=1
D_13=(1/(k+1))*z_13^2*(H[13,13]/(1-H[13,13]))
D_13
```
